## –†–∞—Å—á—ë—Ç –ø—Ä–∏–±—ã–ª–∏ –∏ —Ä–∏—Å–∫–æ–≤ 

```python
# –ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å—á—ë—Ç–∞ –ø—Ä–∏–±—ã–ª–∏ –ø–æ —É—á–∞—Å—Ç–∫—É:
def profit_counter(predictions, target, count_of_good_wells = COUNT_OF_SELECTED_WELLS, barrel_price = 450000):
    target_good = pd.Series(target).sort_values(ascending=False).head(count_of_good_wells)
    predictions_good = pd.Series(predictions).sort_values(ascending=False).head(count_of_good_wells)
    return (predictions_good.sum() * barrel_price) - BUDGET, (target_good.sum() * barrel_price) - BUDGET
```

```python
# –ü—Ä–æ–≤–µ–¥—ë–º —Ä–∞—Å—á—ë—Ç –ø—Ä–∏–±—ã–ª–∏ —Å –ø–æ–º–æ—â—å—é —Ñ—É–Ω–∫—Ü–∏–∏ profit_counter:
predicted_profit_rg_1, target_profit_rg_1 = profit_counter(predictions_1, target_valid_rg_1)
predicted_profit_rg_2, target_profit_rg_2 = profit_counter(predictions_2, target_valid_rg_2)
predicted_profit_rg_3, target_profit_rg_3 = profit_counter(predictions_3, target_valid_rg_3)
print('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–∞—è –ø—Ä–∏–±—ã–ª—å —Å –ø–µ—Ä–≤–æ–≥–æ —É—á–∞—Å—Ç–∫–∞',predicted_profit_rg_1)
print('–§–∞–∫—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–∏–±—ã–ª—å —Å –ø–µ—Ä–≤–æ–≥–æ —É—á–∞—Å—Ç–∫–∞',target_profit_rg_1)
print('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–∞—è –ø—Ä–∏–±—ã–ª—å —Å–æ –≤—Ç–æ—Ä–æ–≥–æ —É—á–∞—Å—Ç–∫–∞',predicted_profit_rg_2)
print('–§–∞–∫—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–∏–±—ã–ª—å —Å–æ –≤—Ç–æ—Ä–æ–≥–æ —É—á–∞—Å—Ç–∫–∞',target_profit_rg_2)
print('–ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω–∞—è –ø—Ä–∏–±—ã–ª—å —Å —Ç—Ä–µ—Ç—å–µ–≥–æ —É—á–∞—Å—Ç–∫–∞',predicted_profit_rg_3)
print('–§–∞–∫—Ç–∏—á–µ—Å–∫–∞—è –ø—Ä–∏–±—ã–ª—å —Å —Ç—Ä–µ—Ç—å–µ–≥–æ —É—á–∞—Å—Ç–∫–∞',target_profit_rg_3)
```

–ö–∞–∫ –≤–∏–¥–Ω–æ –º–æ–¥–µ–ª—å —Å–∏–ª—å–Ω–æ –æ—à–∏–±–∞–µ—Ç—Å—è –≤ –ø–µ—Ä–≤–æ–º –∏ —Ç—Ä–µ—Ç—å–µ–º —Ä–µ–≥–∏–æ–Ω–µ, –≤–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ –ø–æ–ª—É—á–∞–µ—Ç—Å—è –¥–æ–≤–æ–ª—å–Ω–æ —Ç–æ—á–Ω—ã–º

<div class="alert alert-block alert-success"><b>–ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ </b>
    
<b>–£—Å–ø–µ—Ö:</b><br>
üëç –ò —Ä–∞–∑–º–µ—Ä –æ—à–∏–±–∫–∏ –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ —ç—Ç–æ –ø–æ–¥—Ç–≤–µ—Ä–∂–¥–∞–µ—Ç.
</div>

<font size=8> –†–∞—Å—á—ë—Ç —Ä–∏—Å–∫–æ–≤ <font>

```python
def revenue(target, probs, count):
    probs_sorted = pd.Series(probs).sort_values(ascending=False)
    selected = target[probs_sorted.index][:count]
    return (450000 * sum(selected) - BUDGET)

def bootstrap(target, predictions):
    state = np.random.RandomState(12345)
    target = target.reset_index(drop=True)
    predictions = pd.Series(predictions)
    values = []
    counter = 0
    
    for i in range(1000):
        target_subsample = target.sample(n=COUNT_OF_WELLS, replace=True, random_state=state)
        probs_subsample = predictions[target_subsample.index]
        income = revenue(target_subsample, probs_subsample, COUNT_OF_SELECTED_WELLS)
        if income < 0: 
            counter += 1
        values.append(income)
    values = pd.Series(values)
    lower = values.quantile(0.025)
    upper = values.quantile(0.975)
    mean_values = values.mean()
    return (mean_values, counter / len(values), lower, upper)
    
```

–ü–æ—á–µ–º—É –∏–º–µ–Ω–Ω–æ —Ç–∞–∫. –í—Å–ø–æ–º–Ω–∏–º, —á—Ç–æ –≤—ã–±–æ—Ä–∫–∞ - —ç—Ç–æ –Ω–µ–∫–æ—Ç–æ—Ä–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –æ –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏, –µ–µ —á–∞—Å—Ç—å. –ó–Ω–∞—á–∏—Ç, —á—Ç–æ –≥–¥–µ-—Ç–æ –≤ —Ä–µ–≥–∏–æ–Ω–µ —Å—É—â–µ—Å—Ç–≤—É—é—Ç –∏ –¥—Ä—É–≥–∏–µ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—ã–µ —Ç–æ—á–∫–∏ –¥–æ–±—ã—á–∏ –Ω–µ—Ñ—Ç–∏, –æ –∫–æ—Ç–æ—Ä—ã—Ö –º—ã –º–æ–∂–µ–º –Ω–µ –∑–Ω–∞—Ç—å, –Ω–æ –æ–Ω–∏ —è–≤–ª—è—é—Ç—Å—è —á–∞—Å—Ç—å—é —ç—Ç–æ–π –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏. –í —ç—Ç–æ–º —Å–ª—É—á–∞–µ –ø–æ–≤—Ç–æ—Ä—è—é—â–∏–µ—Å—è –Ω–∞–±–ª—é–¥–µ–Ω–∏—è –º–æ–∂–Ω–æ —Å—á–∏—Ç–∞—Ç—å —Ä–∞–∑–Ω—ã–º–∏ —Ç–æ—á–∫–∞–º–∏ –¥–æ–±—ã—á–∏ –∏–∑ –≤—Å–µ–π –≥–µ–Ω–µ—Ä–∞–ª—å–Ω–æ–π —Å–æ–≤–æ–∫—É–ø–Ω–æ—Å—Ç–∏ —Å –±–ª–∏–∑–∫–∏–º–∏ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∞–º–∏.
    
</div>
